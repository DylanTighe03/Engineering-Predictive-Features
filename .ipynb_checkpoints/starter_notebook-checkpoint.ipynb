## Step 3: Engineer New Features

### Feature 1: FullBath - [Category]

**Business Justification:**
Baths are one of the most sought out features of houses. Most people tend to have multiple bathrooms for each member in the house and also some extra for guests. So this is an important feature that needed to be evaluated to preict the housing price and demand.

````python
# TODO: Create your first engineered feat
df['total_bathrooms'] = df['FullBath'] + 0.5 * df['HalfBath']
````

### Feature 2: Quality weighted living room - [Category]

**Business Justification:**
This feature combines the overall quality of the most important areas for a house. By evauluating the size and quality will help list the houses feature of the living area.

````python
# TODO: Create your second engineered feature
df['qual_living_area'] = df['OverallQual'] * df['GrLivArea']
````

### Feature 3: Lot Effeciency - [Category]

**Business Justification:**
Will tell how much of the land is actually livable. This will allow potential buyers to understand the actual aount of HOUSE they are buying and not lumping in land area such as backyard or side yards.

````python
# TODO: Create your third engineered feature
df['lot_efficiency'] = df['GrLivArea'] / df['LotArea']
````

### Feature 4: Garage to living space ratio - [Category]

**Business Justification:**
Shows the amounut of garage space and living area within the house. The garage is usually only marked as how many cars it can fit which is a very important feature when evauluating a house.

````python
# TODO: Create your fourth engineered feature
df['garage_to_living_ratio'] = df['GarageArea'] / df['GrLivArea']
````

### Feature 5: Quality per lot area - [Category]

**Business Justification:**
This reinforces the Lot efficiency by rating the quality of the area and not just the size. This will help show the overall size and amount you are paying for. 

````python
# TODO: Create your fifth engineered feature
df['quality_per_lot'] = df['OverallQual'] / df['LotArea']
````

## Step 4: Train Model with Engineered Features

````python
# Create feature list combining baseline + engineered features
# TODO: List all your engineered feature names
engineered_features = [
    'total_bathrooms',
    'qual_living_area',
    'garage_to_living_ratio',
    'quality_per_lot'
]

# Combine baseline and engineered features
all_features = baseline_features + engineered_features

# TODO: Create X_engineered with all features
# Remember to handle missing values and encode categoricals
X_engineered = df[all_features].copy()
X_engineered = X_engineered.fillna(X_engineered.median(numeric_only=True))  
X_engineered = X_engineered.fillna('None') 

X_engineered = pd.get_dummies(X_engineered, drop_first=True)



print(f"Total features in engineered model: {len(all_features)}")
print(f"New engineered features: {len(engineered_features)}")
````

````python
# Split data (use same random_state for fair comparison)
# TODO: Split X_engineered and y
X_train_eng, X_test_eng, y_train_eng, y_test_eng = train_test_split(
    X_engineered, y, test_size=0.2, random_state=42
)


# Train model with engineered features
# TODO: Train RandomForestRegressor(n_estimators=100, random_state=42)
engineered_model = RandomForestRegressor(n_estimators=100, random_state=42)
engineered_model.fit(X_train_eng, y_train_eng)  # Replace with trained model

# Make predictions
# TODO: Generate predictions on test set
engineered_predictions = engineered_model.predict(X_test_eng)

# Calculate metrics
# TODO: Calculate R² and RMSE
engineered_r2 = r2_score(y_test_eng, engineered_predictions)
engineered_rmse = np.sqrt(mean_squared_error(y_test_eng, engineered_predictions)) # Replace with np.sqrt(mean_squared_error())

print("\n" + "="*80)
print("ENGINEERED MODEL RESULTS")
print("="*80)
print(f"R² Score: {engineered_r2 if engineered_r2 is not None else 'Not calculated'}")
print(f"RMSE: ${engineered_rmse:,.2f}" if engineered_rmse is not None else "RMSE: Not calculated")
print("="*80)
````
